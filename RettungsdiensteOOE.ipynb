# Rettungsdienste OÖ Grafiken
## Motivation
Im Jahr 2022 wurden, nach einigen (teils anonymen) Leserbriefen durch Sanitäter, Defizite der rettungsdienstlichen Versorgung in Oberösterreich in den Medien aufgezeigt.

Deshalb wurde durch die NEOS eine dringliche Anfrage zum Thema Rettungsdienste in Oberösterreich in den oberösterreichischen Landtag eingebracht.
Beantwortet wurde die Anfrage durch Landeshauptmann-Stellvertreterin Mag. Christine Haberlander.

Die hier visualisierten Daten stammen aus der Beantwortung der dringlichen Anfrage.

Link Anfrage: https://www.neos.eu/_Resources/Persistent/2b873f686789cba40fb602890f73ce10a344d236/Rettungsdienste%20in%20Ober%C3%B6sterreich.pdf

Link Beantwortung: https://www.neos.eu/_Resources/Persistent/ca7cae0f5901889a7687cdd85777e07836d1834b/Rettungsdienste%20in%20Ober%C3%B6sterreich-3.pdf

### Abkürzungen
- ASBOÖ: Arbeiter-Samariterbund Österreichs, Landesverband Oberösterreich
- RKOÖ: Österreichisches Rotes Kreuz, Landersverband Oberösterreich
- SEW: Sanitätseinsatzwagen (Ein Fahrzeug für den Krankentransort, niedrig priorisierte Notfälle und hoch priorisierte Notfälle. Bezeichnung nur beim RKOÖ geläufig.)
- RTW: Rettungswagen (Ein Fahrzeug für hoch priorisierte Notfälle. Bezeichnung nur beim ASBOÖ geläufig.)
- NKW: Notfall-Krankenwagen (Ein Fahrzeug für den Krankentransort und niedrig priorisierte Notfälle. Bezeichnung nur beim ASBOÖ geläufig)
- NEF: Notarzt-Einsatzfahrzeug
- RS: Rettungssanitäter
- NFS: Notfallsanitäter
- OÖ: Oberösterreich
## NEF OÖ Kosten
Frage 17:  Wie hoch waren die Gesamtausgaben für das NEF-System

```python
# Zeit-Vektor
x = [2018,2019,2020]
# NEF Kosten OÖ
NAD_cost = [18259418,18988975,19539701]

import numpy as np
import matplotlib.pyplot as plt

# Dividieren um Werte in Mio EUR zu haben
NAD_cost_mio = np.divide(NAD_cost, 1000000)

# Ansicht anpassen
plt.locator_params(axis='y', nbins=3)
plt.locator_params(axis='x', nbins=3)

# Plot
plt.plot(x, NAD_cost_mio)
plt.xlabel('Jahr')
plt.ylabel('Mio EUR')
plt.title('Gesamtkosten NEF-System OÖ EUR 2018-2020')
plt.show()

```

## Anzahl Einsätze der RKOÖ SEWs und NEF
Frage 26: Wie viele Einsätze der Notfallrettung wurden im Zeitraum 2017-2022 durch die zentrale Landesleitstelle pro Jahr an das Rote Kreuz Oberösterreich vergeben, aufgeschlüsselt nach RTW, KTW, NEF.

Anmerkung: Das RKOÖ trennt nicht zwischen RTW und KTW sondern fasst diese zusammen als SEW (Sanitätseinsatzwagen)

```python
# Zeit-Vektor
x = [2017,2018,2019,2020,2021]
# SEW und NEF Einsätze
SEW_einsaetze = [61051,61771,65645,63529,66942]
NEF_einsaetze = [27952,27384,27960,25569,26998]

#Ansicht anpassen
plt.locator_params(axis='y', nbins=5)
plt.locator_params(axis='x', nbins=5)

# Plot SEW Einsätze
plt.plot(x, SEW_einsaetze)
plt.xlabel('Jahr')
plt.ylabel('Einsätze')
plt.title('Einsätze RKOÖ SEW')

plt.show()

# Plot NEF Einsätze
plt.locator_params(axis='y', nbins=5)
plt.locator_params(axis='x', nbins=5)
plt.plot(x, NEF_einsaetze)
plt.xlabel('Jahr')
plt.ylabel('Einsätze')
plt.title('Einsätze NEF')
plt.show()

```

## Einsätze der ASB RTWs und NKWs
Frage 27: Wie viele Einsätze der Notfallrettung wurden im Zeitraum 2017-2022 durch die zentrale Landesleitstelle pro Jahr an den Arbeiter Samariterbund Oberösterreich vergeben, aufgeschlüsselt nach RTW, KTW, NEF?

Anmerkung: Der ASB vergibt nicht direkt Einsätze an den eigenen NEF3, weshalb der Samariterbund diese Daten nicht bereitstellen kann. 

```python
# Einsätze ASBOÖ RTW und NKW
RTW_einsaetze = [4652,4364,4234,4686,5322]
NKW_einsaetze = [2198,2401,2682,1921,1456]
#Einsätze RTW und NKW gemeinsam
ASB_einsaetze = np.array(RTW_einsaetze) + np.array(NKW_einsaetze)

#Ansicht anpassen
plt.locator_params(axis='y', nbins=5)
plt.locator_params(axis='x', nbins=5)

# Plot RTW Einsätze
plt.plot(x, RTW_einsaetze)
plt.xlabel('Jahr')
plt.ylabel('Einsätze')
plt.title('Einsätze RTW ASBOÖ')
plt.show()

# Plot NKW Einsätze
plt.locator_params(axis='y', nbins=5)
plt.locator_params(axis='x', nbins=5)
plt.plot(x, NKW_einsaetze)
plt.xlabel('Jahr')
plt.ylabel('Einsätze')
plt.title('Einsätze NKW ASBOÖ')
plt.show()
```

## Vergleich Einsatzzahlen RKOÖ und ASBOÖ

```python
#Division aller RK SEW Einsätze von allen Jahren durch die Anzahl aller ASB Einsätze zu 2017
divider_2017 = SEW_einsaetze[0]/ASB_einsaetze[0]
divider_2021 = SEW_einsaetze[4]/ASB_einsaetze[4]
print('Das gesamte Rote Kreuz OÖ hatte 2017 um %0.2f mehr Einsätze der ASBOÖ' % (divider_2017))
print('Das gesamte Rote Kreuz OÖ hatte 2021 um %0.2f mehr Einsätze der ASBOÖ' % (divider_2021))
SEW_einsaetze_scaled = np.divide(SEW_einsaetze, divider_2017)

# Ansicht anpassen
plt.locator_params(axis='y', nbins=5)
plt.locator_params(axis='x', nbins=5)
 
# Plot RK
plt.plot(x, SEW_einsaetze_scaled, label="RK")
#plt.xlabel('Jahr')
#plt.ylabel('Einsätze')
#plt.title('Einsätze SEW skaliert')

# Plot ASB
plt.plot(x, ASB_einsaetze, label="ASB")
plt.xlabel('Jahr')
plt.ylabel('Einsätze')
plt.title('Einsätze ASBOÖ und RKOÖ skaliert')
plt.legend(loc="upper left")
plt.show()

```
